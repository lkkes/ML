```python
import argparse

arg_parser = argparse.ArgumentParser()
arg_parser.add_argument('-s', '--srcfile', help='Input file name', required=True)
arg_parser.add_argument('-t', '--tgtfile', help='Input file name', required=True)
arg_parser.add_argument('--lowercase', help='Case independent search', required=False)
args = arg_parser.parse_args()

#args.infile
#args.lowercase

srcfile = open(args.srcfile, encoding='utf8')
srcout = open(args.srcfile + '.clean.txt', mode='w', encoding='utf8')
tgtfile = open(args.tgtfile, encoding='utf8')
tgtout = open(args.tgtfile + '.clean.txt', mode='w', encoding='utf8')
lowercase = args.lowercase

lines_dedup = set()

for srcline in srcfile:
  tgtline = tgtfile.readline()
  srcline = srcline.strip()
  tgtline = tgtline.strip()
  if (srcline, tgtline) not in lines_dedup:
    lines_dedup.add((srcline,tgtline))
    srcout.write(srcline + '\n')
    tgtout.write(tgtline + '\n')

srcfile.close()
tgtfile.close()
srcout.close()
tgtout.close()
```

```python
import sys

inp_en = open(sys.argv[1], encoding='utf8')
inp_ru = open(sys.argv[2], encoding='utf8')

out_en = open(sys.argv[1] + '.clean.cleaned.txt', mode='w', encoding='utf8')
out_ru = open(sys.argv[2] + '.clean.cleaned.txt', mode='w', encoding='utf8')

logfile = open('len_log.txt', mode='w', encoding='utf8')

for line_en in inp_en:
  line_ru = inp_ru.readline()
  line_en = line_en.strip()
  line_ru = line_ru.strip()

  alpha_counter_en = 0
  alpha_counter_ru = 0
  for c in line_en:
    if c.isalpha():
      alpha_counter_en += 1
  for c in line_ru:
    if c.isalpha():
      alpha_counter_ru += 1

  if (len(line_en.split()) < 2) or (len(line_ru.split()) < 2):
    logfile.write('TOO SHORT\n')
    logfile.write(line_en + '\n' + line_ru + '\n\n')
  elif (len(line_en.split()) / len(line_ru.split()) >= 3) or (len(line_ru.split()) / len(line_en.split()) >= 3):
    logfile.write('GREAT DIFF\n')
    logfile.write(line_en + '\n' + line_ru + '\n\n')
  elif (alpha_counter_en / len(line_en) <= 0.8) or (alpha_counter_ru / len(line_ru) <= 0.8):
    logfile.write('NO ALPHA\n')
    logfile.write(line_en + '\n' + line_ru + '\n\n')
  else:
    out_en.write(line_en + '\n')
    out_ru.write(line_ru + '\n')

inp_en.close()
inp_ru.close()
out_en.close()
out_ru.close()
logfile.close()
```

```python
import sys

inp_src = open(sys.argv[1], encoding='utf8')
inp_tgt = open(sys.argv[2], encoding='utf8')
ld_src = open(sys.argv[3], encoding='utf8')
ld_tgt = open(sys.argv[4], encoding='utf8')
inp_l = sys.argv[5]
tgt_l = sys.argv[6]

out_src = open(sys.argv[1] + '.clean.txt', mode='w', encoding='utf8')
out_tgt = open(sys.argv[2] + '.clean.txt', mode='w', encoding='utf8')
log = open('ld.log', mode='w', encoding='utf8')

slavic_langs = ('ru', 'uk', 'be', 'sr', 'bg', 'mn', 'mk', 'kk', 'ba', 'tt', 'cv')

for src_line in inp_src:
  tgt_line = inp_tgt.readline()
  ld_src_line = ld_src.readline()
  ld_tgt_line = ld_tgt.readline()
  lds = ld_src_line.split()[0]
  ldt = ld_tgt_line.split()[0]
  if lds != inp_l or ldt != tgt_l:
    if inp_l in slavic_langs and lds in slavic_langs:
      out_src.write(src_line)
      out_tgt.write(tgt_line)
    elif tgt_l in slavic_langs and ldt in slavic_langs:
      out_src.write(src_line)
      out_tgt.write(tgt_line)
    else:
      log.write(lds + '\t' + src_line)
      log.write(ldt + '\t' + tgt_line + '\n')
  else:
    out_src.write(src_line)
    out_tgt.write(tgt_line)

inp_src.close()
inp_tgt.close()
ld_src.close()
ld_tgt.close()
out_src.close()
out_tgt.close()
log.close()
```

```python
scored_file = open('corpus.sorted', encoding='utf8')
threshold = 0.5

src_file = open('corpus.en-ru.selected.en', mode='w', encoding='utf8')
tgt_file = open('corpus.en-ru.selected.ru', mode='w', encoding='utf8')
log = open('bicleaner.log', mode='w', encoding='utf8')

for line in scored_file:
  try:
    score, sline, tline = line.strip().split('\t')
    score = float(score)
    if score >= threshold:
      src_file.write(sline + '\n')
      tgt_file.write(tline + '\n')
    else:
      log.write(line)
  except ValueError:
    print(line)

scored_file.close()
src_file.close()
tgt_file.close()
```

```python
import sys, random

num_lines = int(sys.argv[1])
num_rand_lines = int(sys.argv[2])
rest_flag = int(sys.argv[3])

indices = set(random.sample(range(num_lines), num_rand_lines))

for filename in sys.argv[4:]:
  inp = open(filename, encoding='utf8')
  out = open(filename + '.dev' + str(num_rand_lines), mode='w', encoding='utf8')
  if rest_flag == 1:
    out_rest = open(filename + '.rest' + str(num_rand_lines), mode='w', encoding='utf8')

  for i, line in enumerate(inp):
    if i in indices:
      out.write(line)
    else:
      if rest_flag == 1:
        out_rest.write(line)
  inp.close()
  out.close()
  if rest_flag == 1:
    out_rest.close()
```
